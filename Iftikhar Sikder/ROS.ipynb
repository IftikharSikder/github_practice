```python
#importing dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from imblearn.over_sampling import RandomOverSampler
from collections import Counter


# Step 1: Importing dataset
data = pd.read_csv("C:/Users/eftys/OneDrive/Desktop/Research Paper/Sahriar/Feature selected.csv")

# Step 2: Visualize the class distribution before ROS
class_distribution_before = data['status'].value_counts()
print("Class Distribution Before ROS:")
print(class_distribution_before)

plt.figure(figsize=(6, 4))
class_distribution_before.plot(kind='bar', color='#5E9CA0')
plt.title('Class Distribution Before ROS')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Step 3: Apply ROS to balance the classes
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(data.drop('status', axis=1), data['status'])

# Concatenate resampled features and target variable
resampled_data = pd.concat([X_resampled, y_resampled], axis=1)

# Step 4: Visualize the class distribution after ROS
class_distribution_after = pd.Series(y_resampled).value_counts()
print("\nClass Distribution After ROS:")
print(class_distribution_after)

plt.figure(figsize=(6, 4))
class_distribution_after.plot(kind='bar', color='#5E9CA0')
plt.title('Class Distribution After ROS')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Step 5: Save the resampled dataset to a CSV file
resampled_data.to_csv("C:/Users/eftys/OneDrive/Desktop/Research Paper/Sahriar/after_ros.csv", index=False)

```
